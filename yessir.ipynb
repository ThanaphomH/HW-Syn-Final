```python
from PIL import Image
import texttoimage

def image_to_ascii(txt, output_width=100, output_height=50):
    """
    Convert an image to ASCII art using only '0' and '1'.
    
    Parameters:
        image_path (str): Path to the input image.
        output_width (int): Width of the ASCII output.
        output_height (int): Height of the ASCII output.
    
    Returns:
        str: ASCII art string using '0' and '1'.
    """
    texttoimage.convert(txt, image_file='test.png', font_name="Times New Roman.ttf", font_size=120, color='white')
    # Load the image
    img = Image.open('test.png')
    
    # Resize the image to the specified dimensions
    img = img.resize((output_width, output_height))
    
    # Convert the image to grayscale
    img = img.convert("L")
    
    # Define the mapping from grayscale to '0' or '1'
    threshold = 128  # Midpoint of grayscale range (0-255)
    ascii_art = []

    for y in range(output_height):
        cur = []
        for x in range(output_width):
            pixel = img.getpixel((x, y))
            cur.append(0 if pixel < threshold else 1)
        ascii_art.append(cur)
    
    return ascii_art

# Example usage
image_path = "test.png"  # Replace with your image file path
output_width = 8  # Specify desired width
output_height = 16  # Specify desired height

# ascii_art = image_to_ascii("ก", output_width, output_height)

# import matplotlib.pyplot as plt
# plt.imshow(ascii_art, cmap='gray')
# plt.axis('off')  # Turn off axis labels
# plt.title("Monochrome Picture")
# plt.show()
```
